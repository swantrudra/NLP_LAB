## 2. Synonymy Example

```python

word = "happy"
synonyms = set()

for syn in wn.synsets(word):
    for lemma in syn.lemmas():
        synonyms.add(lemma.name())

print("Synonyms of", word, ":", synonyms)
```

## 3. Antonymy Example

```python

word = "hot"
antonyms = set()

for syn in wn.synsets(word):
    for lemma in syn.lemmas():
        if lemma.antonyms():
            antonyms.add(lemma.antonyms()[0].name())

print("Antonyms of", word, ":", antonyms)
```

## 4. Hypernymy Example

```python

word = "dog"
synsets = wn.synsets(word)

if synsets:
    hypernyms = synsets[0].hypernyms()
    print("Hypernyms of", word, ":")
    for h in hypernyms:
        print(" -", h.name())
```

## 5. Hyponymy Example

```python

word = "animal"
synsets = wn.synsets(word)

if synsets:
    hyponyms = synsets[0].hyponyms()
    print("Hyponyms of", word, ":")
    for h in hyponyms[:10]:
        print(" -", h.name())
```

```python
sentence = "The dog is very happy today"
tokens = sentence.lower().split()

print("Sentence:", sentence)
print("Tokens:", tokens)
```

```python

from nltk.corpus import wordnet as wn

for token in tokens:
    synsets = wn.synsets(token)
    if synsets:
        synonyms = {lemma.name() for lemma in synsets[0].lemmas()}
        print(f"Synonyms of '{token}':", synonyms)
```

```python

for token in tokens:
    for syn in wn.synsets(token):
        for lemma in syn.lemmas():
            if lemma.antonyms():
                print(f"Antonym of '{token}':", lemma.antonyms()[0].name())
```

```python

for token in tokens:
    synsets = wn.synsets(token)
    if synsets:
        hypernyms = synsets[0].hypernyms()
        if hypernyms:
            print(f"Hypernym of '{token}':", hypernyms[0].name())
```
